```python
#concatination in pandas
#concat() is used to combine multiple dataframes
```

```python
import pandas as pd
```

```python
indian_weather=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
print(indian_weather)
```

```python
paris_weather=pd.DataFrame({
    "city":["paris","london"],
    "temperature":[42,30],
    "humidity":[68,75]
})
print(paris_weather)
```

```python
df=pd.concat([indian_weather,paris_weather])
print(df)
```

```python
df=pd.concat([indian_weather,paris_weather],ignore_index=True)
print(df)
```

```python
df=pd.concat([indian_weather,paris_weather],keys=["ind","paris"])
print(df)
```

```python
df=pd.concat([indian_weather,paris_weather],axis=1)
print(df)
```
