```python
import pandas as pd
```

```python
student_df=pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/students.csv")
print(student_df)
```

```python
marks_df=pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/marks.csv")
print( marks_df)
```

```python
mergedf=pd.merge(student_df,marks_df,on="student_id",how="inner")##############inner merge commomn data only
print(mergedf)
```

```python
##########left merge data data on leftside based
mergedf=pd.merge(student_df,marks_df,on="student_id",how="left")
print(mergedf)
```

```python
mergedf=pd.merge(student_df,marks_df,on="student_id",how="right")
print(mergedf)
```

```python
mergedf=pd.merge(student_df,marks_df,on="student_id",how="outer")
print(mergedf)
```
